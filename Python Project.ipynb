<h2 style="text-align:left;color:darkblue;">Imputing Missing Value</h2>

```python
num_cols = ['BMI', 'Cholesterol_Level', 'Systolic_BP', 'Diastolic_BP', 'Resting_Heart_Rate', 'HbA1c', 'Medication_Adherence_Percentage']
for col in num_cols:
    df[col] = df[col].fillna(df[col].mean()) # Numbers column-->fill using Mean

# 2. Text column---> Fill using Mode
cat_cols = ['Smoking_Status', 'Alcohol_Consumption', 'Physical_Activity_Level']
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Missing values after Imputation:")
print(df.isnull().sum())
```

```python
mapping = {0: 'No', 1: 'Yes'}

cols_to_map = ['Diabetes', 'Hypertension', 'Chronic_Kidney_Disease', 'Liver_Disease', 'Family_History_CVD']

for col in cols_to_map:
    df[col] = df[col].map(mapping)
```

<h2 style="text-align:left;color:darkblue;">Detection of outliers & Removing of Outliers using IQR</h2>

```python
columns_to_check = ['Age', 'BMI', 'Cholesterol_Level', 'Systolic_BP', 'Diastolic_BP','HbA1c', 'Resting_Heart_Rate']

# FIND Q1, Q3
Q1 = df[columns_to_check].quantile(0.25)
Q3 = df[columns_to_check].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# CHECK, OUTLIERS CONDITION
outliers_condition = (
    (df[columns_to_check] < lower_limit)
    | (df[columns_to_check] > upper_limit)
).any(axis=1)

outliers = df[outliers_condition]
print("Count of Outliers:", outliers.shape[0])
outliers
```

```python
# Filter using lower_limit, upper_limit 
columns_to_check =['Age', 'BMI', 'Cholesterol_Level', 'Systolic_BP', 'Diastolic_BP','HbA1c', 'Resting_Heart_Rate']

# Non_outliers move to another data frame
df_clean = df.copy()

for col in columns_to_check:
    df_clean = df_clean[(df_clean[col] >= lower_limit[col]) & (df_clean[col] <= upper_limit[col])]

print("Shape after removing outliers (using IQR):", df_clean.shape)
```

<h2 style="text-align:left;color:darkblue;">Feature Engineering: BMI Categorization</h2>

```python
# Defines the BMI category
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal Weight'
    elif 25 <= bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

# create a new column
df_clean['BMI_Category'] = df_clean['BMI'].apply(categorize_bmi)

# Checking the first few rows
print(df_clean[['BMI', 'BMI_Category']].head())
```

```python
diseases = ["Diabetes", "Hypertension", "Chronic_Kidney_Disease", "Liver_Disease"]
df_melted = df_clean.melt(
    id_vars=["Mortality_Risk_Percentage"],
    value_vars=diseases,
    var_name="Disease",
    value_name="Status",
)

fig = px.histogram(
    df_melted,
    x="Disease",
    y="Mortality_Risk_Percentage",
    color="Status",
    barmode="group",
    histfunc="avg",
    text_auto=".1f",
    title="Average Mortality Risk by Disease Status",
)
fig.update_layout(width=1000, height=600)
fig.update_layout(yaxis_title="Average Mortality Risk Percentage (%)")
fig.show()
```

**💡Insight:** 
* Highest Danger: Diabetes is the most dangerous disease in this chart. People with diabetes have the highest risk of death at 29.2%. Chronic Kidney Disease (CKD) is the second most dangerous at 25.5%.

* Big Jump in Risk: Having a disease makes the risk of death go up a lot. For example, people with Hypertension (high blood pressure) see their risk jump from 7.3% to 23.0%—which is more than 3 times higher!

* Lowest Danger: Liver Disease has the smallest effect. The risk for people without it is 10.1%, and it only goes up a little bit to 11.6% for people who have it.
